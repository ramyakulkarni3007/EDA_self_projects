```python
df['class_int']=pd.Categorical(df['Class']).codes # to convert the class name to integer
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
target='class_int'
y=df[target]
df[target]
```

```python
df_train=df.drop(columns=['Class','class_int'])
```

```python
x=df_train
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
df_scaled = pd.DataFrame(x_scaled)
```

```python
df_scaled
```

```python
# x_scaled.info()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_scaled,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
pred=rfc.predict(x_test)#thats my own value(predicted output)
```

```python
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score, confusion_matrix
```

```python
report = classification_report(y_test, pred)
cm = confusion_matrix(y_test, pred)
rfc_f1 = f1_score(y_test, pred)
rfc_Precision = precision_score(y_test, pred)
rfc_Recall = recall_score(y_test, pred)
```

```python
print(report)
print('Confusion Matrix : \n{}\n'.format(cm))
print("F1 Score         : {:.4}%\n".format(rfc_f1*100))
print("Precision Score  : {:.4}%\n".format(rfc_Precision*100))
print("Recall Score     : {:.4}%".format(rfc_Recall*100))
```
