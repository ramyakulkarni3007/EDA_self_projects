```python
from scipy.stats.mstats import winsorize
import numpy as np

def winsor(x,multiplier=3):
    upper = x.median()+x.std()*multiplier
    
    for limit in np.arange(0.001,0.2,0.001):
        if np.max(winsorize(x,(0,limit)))<upper:
            return limit
        
    return None

xtrain_col = []
for cols in df.columns:
    xtrain_col.append(cols)

    
for i in range(1,len(xtrain_col)):
    df[xtrain_col[i]]= winsorize(df[xtrain_col[i]],(0,winsor(df[xtrain_col[i]]))) #,(0,winsor(df[xtrain_col[i]]))
```
